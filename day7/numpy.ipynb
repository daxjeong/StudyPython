## 리스트와 행렬의 차이점
- 리스트 : 사용자가 막 사용할 수 있도록 만든 구조(입력하기 쉬움)
- 배열 : 계산에 최적화된 행과 열 처리 기법(속도가 빠름, GPU등, 게임 필수)
- 행렬구조에 특화된 numpy 모듈에 대하여 먼저 학습
- 행렬은 반드시 행과 열의 갯수가 인덱싱마다 같아야하고, 연산이 가능함. 자료의 성격이 같아야함.
- 행렬에 특화된 모듈은 numpy임

```python
a = [1,2,3,4,5]
b = [10,20,30,40,50]
val = 3
cc = []
for x1,x2 in zip(a,b):
    cc.append(x1+x2)

cc

# 행렬의 수가 다르면 값이 안나온다는 한계가 있음 => numpy 사용
```

```python
import os
folderName=('./pic/')
file_list = os.listdir('./pic/')
file_list
```

```python
from PIL import Image
# 컬러이미지는 [R, G, B]를 갖고 있고, 각 R,G,B는 0~255값임. 255,0,0 = Red
fileName = folderName+'/'+file_list[0]
img = Image.open(fileName)
# img = Image.open('C:\\Users\\admin\\Desktop\\pic\\1.jpg') # ./ : 현재 작업 폴더 아래에(./pic/)

# 사이즈 조절 : 여러개의 이미지를 처리하기 위해서는 사이즈(=행렬)가 같아야 함 => 행렬 연산이 가능해짐
img = img.resize((200,200))
img = img.rotate(35)
# img = img.resize((100,200)) # 너비 100, 높이 200 / 행은 200, 열은 100
img # 불러온 이미지를 보는 것
```

```python
imgArray = np.array(img) # 이미지가 행과 열로 구성된 자료구조
imgArray.shape # np.shape(img)
```

- BroadCasting : 모양이 다른 배열들 간의 연산이 어떤 조건을 만족했을 때 가능해지도록 배열을 자동적으로 변환하는 것

```python
plt.imshow(imgArray[:20])
imgArray[:40] = imgArray[:40]*0 # BroadCast
plt.imshow(imgArray)
```

```python
re = imgArray*[1,1,2]
plt.imshow(re)
```

```python
val = [[1,0,0],[0,1,0],[0,0,1]] # [1,0,0] : 그린과 블루 제거 # [0,0,0] : 블랙
for i in val:
    re=imgArray * i
    plt.imshow(re) # 여러 개 중에 마지막 사진만 나옴
    plt.show() # 모든 사진 다 나옴
```
